# CONNECTING MODEL

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, r2_score

# 1. Dataset Generator
def generate_data(num_samples=100000, low=0.001, high=1000.0):
    col1 = np.random.uniform(low, high, num_samples)
    col2 = np.random.uniform(low, high, num_samples)
    ops = np.random.choice(['+', '-', '*', '/'], num_samples)
    results = []

    for x, y, op in zip(col1, col2, ops):
        if op == '+':
            res = x + y
        elif op == '-':
            res = x - y
        elif op == '*':
            res = x * y
        elif op == '/':
            res = x / y if abs(y) > 1e-6 else np.nan
        results.append(res)

    df = pd.DataFrame({
        'Col1': col1,
        'Operator': ops,
        'Col2': col2,
        'Result': results
    }).dropna()
    return df

# 2. Train model per operator
def train_model(df_op, operator):
    df = df_op.copy()
    X = df[['Col1', 'Col2']].copy()
    y = df['Result'].copy()

    # Transform for '*' and '/' to improve regression
    if operator in ['*', '/']:
        X = np.log1p(X)
        y = np.log1p(y)

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    model = Sequential([
        Input(shape=(2,)),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(1, activation='linear')
    ])

    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    model.fit(X_train, y_train, validation_split=0.1, epochs=50, batch_size=32, verbose=0)

    y_pred = model.predict(X_test).flatten()

    # Reverse transformation if needed
    if operator in ['*', '/']:
        y_pred = np.expm1(y_pred)
        y_test = np.expm1(y_test)

    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    print(f"\nOperator: {operator}")
    print(f"R² score: {r2:.4f}")
    print(f"MAE: {mae:.4f}")
    if r2 < 0.96:
        print(f"⚠ Warning: Model for '{operator}' has R² < 0.96")

    return model, scaler

# 3. Generate and train
df = generate_data(100000)

models = {}
scalers = {}

for op in ['+', '-', '*', '/']:
    df_op = df[df['Operator'] == op]
    model, scaler = train_model(df_op, op)
    models[op] = model
    scalers[op] = scaler

print("\n✅ All models trained successfully.")

# 4. Inference loop
while True:
    try:
        print("\n--- Arithmetic Predictor ---")
        val1 = float(input("Enter first number: "))
        op = input("Operator (+, -, *, /): ").strip()
        val2 = float(input("Enter second number: "))

        if op not in models:
            print("Invalid operator.")
            continue
        if op == '/' and abs(val2) < 1e-6:
            print("⚠ Division by zero not allowed.")
            continue

        X_input = pd.DataFrame([[val1, val2]], columns=['Col1', 'Col2'])
        if op in ['*', '/']:
            X_input = np.log1p(X_input)

        X_scaled = scalers[op].transform(X_input)
        pred = models[op].predict(X_scaled).flatten()[0]
        if op in ['*', '/']:
            pred = np.expm1(pred)

        print(f"Predicted: {val1} {op} {val2} = {pred:.4f}")
    except KeyboardInterrupt:
        print("\nExiting...")
        break
    except Exception as e:
        print(f"Error: {e}")
```
